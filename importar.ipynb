```python
# Importar librerias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Leer excel de GitHub
data=pd.read_excel("/content/Colab.TesisMESAP/Base1.xlsx")
```

```python
# Mostrar datos
data.head()
```

```python
df = pd.DataFrame(data)

# Realizar el análisis de frecuencias
frequency_analysis = df.apply(pd.Series.value_counts).fillna(0).astype(int)

# Mostrar el análisis de frecuencias
print(frequency_analysis)

# Crear un diagrama de cajas (boxplot)
plt.figure(figsize=(12, 8))
sns.boxplot(data=df, orient='h')
plt.title('Diagrama de Cajas')
plt.xlabel('Valores')
plt.ylabel('Parámetros')
plt.show()
```
